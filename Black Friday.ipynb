## Analysis for population mean

```python
no_of_samples = 1000
size_of_each_sample = 15000

means_of_population_sample = []

for i in range(no_of_samples):
    means_of_population_sample.append(np.mean(df['Purchase'].sample(n = no_of_samples)))
```

```python
sns.histplot(means_of_population_sample,kde=True)
```

```python
import scipy.stats as st

mean_of_mean_of_samples = np.mean(means_of_population_sample)
std_of_mean_of_samples = np.std(means_of_population_sample)

print("mean of mean of samples :",mean_of_mean_of_samples)
print("std of mean of samples :",std_of_mean_of_samples,"\n")

def confidence_interval(mean,std,n,alpha):
    
    ppf = round(st.norm.ppf(1-((alpha/2)/100)),2)
    ppf_X_std_of_mean_of_samples = round(ppf*std,2)
    sqrt_n = round(np.sqrt(n),2)
    
    print(100-alpha,"% Confidence Interval")
    print("Lower bound of population mean :",mean-(ppf_X_std_of_mean_of_samples/sqrt_n))
    print("Upper bound of population mean :",mean+(ppf_X_std_of_mean_of_samples/sqrt_n))
    print("--------------------------------------------------------------\n")

for i in [10,5,1]:
    confidence_interval(mean_of_mean_of_samples,std_of_mean_of_samples,no_of_samples,i)
```

As confidence increases confidence interval gets wider and wider
